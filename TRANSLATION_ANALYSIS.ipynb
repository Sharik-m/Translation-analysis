```python
from IPython import get_ipython
from IPython.display import display
# %%
# import all lib.. in python

#all model ml

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import random
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB






# %%
# step 2
#  importing data set

url = 'https://raw.githubusercontent.com/TheiScale/YouTube-Video-Notes/main/Machine%20Learning%2BNLP%20Project%20Language%20Detection/language.csv'


# Read the CSV using pd.read_csv
data_original = pd.read_csv(url)  # Assign the original DataFrame to data_original
data_original.isnull().sum()# data cleaning

# visualization data

data_original['language'].value_counts()



#counvert array all the columns

#create an array

 # check the data types of columns

#object convert to array

# using numpy libraries

x=np.array(data_original['Text'])
y=np.array(data_original['language'])
# %%
data_original

# %%
from sklearn.feature_extraction.text import CountVectorizer
# create a CountVectorizer object
vectorizer =CountVectorizer()

#sample text data
data=['love data science ','love machine learning'] # This 'data' is a list

#fit transform the data

vectorized_data=vectorizer.fit_transform(data)


#get the vocubulary(unique words)
print(vectorizer.get_feature_names_out())


#convert the result to an array
print(vectorized_data.toarray())


# %%
from sklearn.model_selection import train_test_split
# %%
from sklearn.naive_bayes import MultinomialNB

# %%


# %%
print(x)


# %%
print(y)


# %%
# cv=CountVectorizer()
# #counvert x-array data numbers.
# #createing ml model
# X=cv.fit_transform(data['Text'])

from sklearn.feature_extraction.text import CountVectorizer

# Create a CountVectorizer instance
cv = CountVectorizer()

# Convert text data into a matrix of token counts
# Use the original DataFrame 'data_original' here
X = cv.fit_transform(data_original['Text'])


# %%
# train test split data
#converting numbers metrix
X_train,X_test , y_train ,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
# %%
# print(X_train)
# %%
#-------matplotlib---- labels


# print(y_test)
# %%
model=MultinomialNB()
# %%
model.fit(X_train,y_train)
# %%
model.score(X_test,y_test)
# %%
# check the out


# second part

print('analysis the result of search')
user=input('Enter a text=')
data=cv.transform([user]).toarray()
output=model.predict(data)
print(output)

# %%

# dash board
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Read the Dataset
url = 'https://raw.githubusercontent.com/TheiScale/YouTube-Video-Notes/main/Machine%20Learning%2BNLP%20Project%20Language%20Detection/language.csv'
data = pd.read_csv(url)

# Step 2: Explore the Dataset
print(data.head())  # Check the first few rows of the dataset

# Step 3: Create a Dashboard with Multiple Charts
# Assuming the dataset has 'language' and 'Text' columns

# Set up the figure and axes for the dashboard
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
fig.suptitle('Language Detection Dashboard', fontsize=16)

# ---Changes start here---
# Bar Chart - Changed 'Language' to 'language' and 'Count' to 'Text'
axs[0, 0].bar(data['language'], data['Text'].str.len(), color='skyblue') # plot str length of Text column
# ---Changes end here---
axs[0, 0].set_title('Language Distribution')
axs[0, 0].set_xlabel('Languages')
axs[0, 0].set_ylabel('Count')
axs[0, 0].tick_params(axis='x', rotation=45)

# ---Changes start here---
# Pie Chart - Changed 'Language' to 'language' and 'Count' to 'Text'
axs[0, 1].pie(data['language'].value_counts(), labels=data['language'].unique(), autopct='%1.1f%%', startangle=140) # plot value counts
# ---Changes end here---
axs[0, 1].set_title('Language Distribution (Pie Chart)')


# ---Changes start here---
# Horizontal Bar Chart - Changed 'Language' to 'language' and 'Count' to 'Text'
axs[1, 0].barh(data['language'], data['Text'].str.len(), color='lightgreen')# plot str length of Text column
# ---Changes end here---
axs[1, 0].set_title('Language Distribution (Horizontal Bar Chart)')
axs[1, 0].set_xlabel('Count')
axs[1, 0].set_ylabel('Languages')

# ---Changes start here---
# Line Chart - Changed 'Language' to 'language' and 'Count' to 'Text'
axs[1, 1].plot(data['language'], data['Text'].str.len(), marker='o', color='orange')# plot str length of Text column
# ---Changes end here---
axs[1, 1].set_title('Language Count Over Languages')
axs[1, 1].set_xlabel('Languages')
axs[1, 1].set_ylabel('Count')
axs[1, 1].tick_params(axis='x', rotation=45)



# Adjust layout
plt.tight_layout(rect=[0, 0.03, 1, 0.95])


plt.show()
```
